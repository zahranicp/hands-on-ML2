```python
# ============================================
# EXERCISE 4: Why Logistic Activation Was Key
# ============================================

print("\n" + "=" * 60)
print("EXERCISE 4: LOGISTIC ACTIVATION FUNCTION")
print("=" * 60)

print("\nüìå QUESTION:")
print("Why was the logistic activation function a key ingredient")
print("in training the first MLPs?")

print("\n‚úÖ ANSWER:")
print("-" * 60)

print("\nüîë WHY LOGISTIC (SIGMOID) WAS CRUCIAL:")

print("\n1. DIFFERENTIABILITY:")
print("   ‚Ä¢ Step function: NOT differentiable")
print("   ‚Ä¢ Sigmoid: Smooth, differentiable everywhere")
print("   ‚Ä¢ Enables gradient-based optimization (backpropagation)")

print("\n2. NON-ZERO GRADIENTS:")
print("   ‚Ä¢ Step function: Gradient = 0 almost everywhere")
print("   ‚Ä¢ Sigmoid: Always has non-zero gradient (except at extremes)")
print("   ‚Ä¢ Allows weights to update via gradient descent")

print("\n3. SMOOTH OUTPUT:")
print("   ‚Ä¢ Step function: Discrete jumps {0, 1}")
print("   ‚Ä¢ Sigmoid: Continuous output (0, 1)")
print("   ‚Ä¢ Better for probability interpretation")

print("\n4. CHAIN RULE COMPATIBILITY:")
print("   ‚Ä¢ Backpropagation requires chain rule")
print("   ‚Ä¢ Sigmoid derivative: œÉ'(z) = œÉ(z)(1 - œÉ(z))")
print("   ‚Ä¢ Simple to compute and propagate gradients")

# Mathematical comparison
print("\nüìê MATHEMATICAL COMPARISON:")
print("-" * 60)
print("Step Function:")
print("  f(z) = {1 if z ‚â• 0, else 0}")
print("  f'(z) = 0 (almost everywhere) ‚ùå")

print("\nLogistic (Sigmoid):")
print("  œÉ(z) = 1 / (1 + e^(-z))")
print("  œÉ'(z) = œÉ(z) √ó (1 - œÉ(z)) ‚úÖ")

# Visualize derivatives
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
x = np.linspace(-6, 6, 1000)

# Plot 1: Functions
ax1 = axes[0]
step = np.where(x >= 0, 1, 0)
sigmoid = 1 / (1 + np.exp(-x))

ax1.plot(x, step, 'r-', linewidth=3, label='Step Function', alpha=0.7)
ax1.plot(x, sigmoid, 'b-', linewidth=3, label='Sigmoid', alpha=0.7)
ax1.axhline(y=0, color='gray', linestyle='--', alpha=0.3)
ax1.axvline(x=0, color='gray', linestyle='--', alpha=0.3)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('z', fontsize=12, weight='bold')
ax1.set_ylabel('f(z)', fontsize=12, weight='bold')
ax1.set_title('Activation Functions', fontsize=14, weight='bold')
ax1.legend(fontsize=11)

# Plot 2: Derivatives
ax2 = axes[1]
step_derivative = np.zeros_like(x)  # Always 0
sigmoid_derivative = sigmoid * (1 - sigmoid)

ax2.plot(x, step_derivative, 'r-', linewidth=3, label="Step' (always 0)", alpha=0.7)
ax2.plot(x, sigmoid_derivative, 'b-', linewidth=3, label="Sigmoid'", alpha=0.7)
ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.3)
ax2.axvline(x=0, color='gray', linestyle='--', alpha=0.3)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('z', fontsize=12, weight='bold')
ax2.set_ylabel("f'(z)", fontsize=12, weight='bold')
ax2.set_title('Derivatives (Gradients)', fontsize=14, weight='bold')
ax2.legend(fontsize=11)

plt.tight_layout()
plt.show()

print("\nüí° HISTORICAL CONTEXT:")
print("-" * 60)
print("‚Ä¢ 1960s: Perceptrons with step functions couldn't learn XOR")
print("‚Ä¢ 1986: Backpropagation + Sigmoid = First successful MLPs")
print("‚Ä¢ Sigmoid enabled gradient descent to work")
print("‚Ä¢ This breakthrough led to the neural network revolution")

print("\n‚ö†Ô∏è MODERN NOTE:")
print("-" * 60)
print("Today, ReLU is preferred over sigmoid for hidden layers")
print("Reasons:")
print("  ‚Ä¢ Faster to compute")
print("  ‚Ä¢ No vanishing gradient problem")
print("  ‚Ä¢ Better performance in deep networks")
print("\nBut sigmoid still used for:")
print("  ‚Ä¢ Output layer (binary classification)")
print("  ‚Ä¢ Gates in LSTMs/GRUs")

print("\n‚úÖ Exercise 4 completed!")
```

```python
# ============================================
# EXERCISE 5: Three Popular Activation Functions
# ============================================

print("\n" + "=" * 60)
print("EXERCISE 5: POPULAR ACTIVATION FUNCTIONS")
print("=" * 60)

print("\nüìå QUESTION:")
print("Name three popular activation functions and draw their graphs")

print("\n‚úÖ ANSWER:")
print("-" * 60)

# Define activation functions
x = np.linspace(-5, 5, 1000)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def sigmoid(x):
    return 1 / (1 + np.exp(-np.clip(x, -500, 500)))

def tanh(x):
    return np.tanh(x)

def swish(x):
    return x * sigmoid(x)

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Plot all activation functions
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

activations = [
    ('ReLU', relu, 'Most popular for hidden layers'),
    ('Leaky ReLU', leaky_relu, 'Fixes dying ReLU problem'),
    ('Sigmoid', sigmoid, 'Output layer (binary classification)'),
    ('Tanh', tanh, 'Better than sigmoid (zero-centered)'),
    ('Swish', swish, 'Smooth, better than ReLU sometimes'),
    ('ELU', elu, 'Smooth negative values, like Leaky ReLU')
]

for idx, (name, func, desc) in enumerate(activations):
    ax = axes[idx // 3, idx % 3]
    y = func(x)
    
    ax.plot(x, y, linewidth=3, color='blue')
    ax.axhline(y=0, color='gray', linestyle='--', alpha=0.5, linewidth=1)
    ax.axvline(x=0, color='gray', linestyle='--', alpha=0.5, linewidth=1)
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('Input (z)', fontsize=11, weight='bold')
    ax.set_ylabel('Output f(z)', fontsize=11, weight='bold')
    ax.set_title(f'{name}\n{desc}', fontsize=12, weight='bold')
    ax.set_ylim(-2, 5)

plt.tight_layout()
plt.show()

print("\nüìä ACTIVATION FUNCTIONS SUMMARY:")
print("=" * 90)

summary_data = {
    'Function': ['ReLU', 'Leaky ReLU', 'Sigmoid', 'Tanh', 'Swish', 'ELU'],
    'Formula': [
        'max(0, z)',
        'max(Œ±z, z) (Œ±=0.01)',
        '1 / (1 + e^(-z))',
        '(e^z - e^(-z)) / (e^z + e^(-z))',
        'z √ó œÉ(z)',
        'z if z‚â•0, Œ±(e^z-1) otherwise'
    ],
    'Range': [
        '[0, ‚àû)',
        '(-‚àû, ‚àû)',
        '(0, 1)',
        '(-1, 1)',
        '(-‚àû, ‚àû)',
        '(-Œ±, ‚àû)'
    ],
    'Pros': [
        'Fast, no vanishing gradient',
        'No dead neurons',
        'Probabilistic output',
        'Zero-centered',
        'Smooth, better gradients',
        'Smooth, better than ReLU'
    ],
    'Cons': [
        'Dead neurons possible',
        'Not zero-centered',
        'Vanishing gradients',
        'Vanishing gradients',
        'Slower than ReLU',
        'Slower (exponential)'
    ],
    'Use Case': [
        'Default for hidden layers',
        'When ReLU dies',
        'Binary output layer',
        'RNNs, LSTMs',
        'Research/experimentation',
        'Alternative to ReLU'
    ]
}

df_activations = pd.DataFrame(summary_data)
print(df_activations.to_string(index=False))
print("=" * 90)

print("\nüí° CHOOSING ACTIVATION FUNCTIONS:")
print("-" * 60)
print("HIDDEN LAYERS:")
print("  1st choice: ReLU (fast, works well)")
print("  2nd choice: Leaky ReLU (if ReLU dies)")
print("  3rd choice: ELU or Swish (research)")

print("\nOUTPUT LAYER:")
print("  Binary classification: Sigmoid")
print("  Multi-class classification: Softmax")
print("  Regression: None (linear)")

print("\n‚úÖ Exercise 5 completed!")
```

```python
# ============================================
# EXERCISE 6: MLP Architecture Calculations
# ============================================

print("\n" + "=" * 60)
print("EXERCISE 6: MLP ARCHITECTURE CALCULATIONS")
print("=" * 60)

print("\nüìå QUESTION:")
print("Given an MLP with:")
print("  ‚Ä¢ 10 input neurons")
print("  ‚Ä¢ 50 output neurons")
print("  ‚Ä¢ 3 hidden layers (50 neurons each)")
print("\nCalculate:")
print("  a) Shape of input matrix X for mini-batch of 4 instances")
print("  b) Shape of hidden layer's weight matrix W_h")
print("  c) Shape of output layer's weight matrix W_o")
print("  d) Shape of network's output Y for mini-batch")
print("  e) Write equation to compute Y")

print("\n‚úÖ ANSWER:")
print("-" * 60)

print("\nüèóÔ∏è NETWORK ARCHITECTURE:")
print("-" * 60)
print("Layer 0 (Input):     10 neurons")
print("Layer 1 (Hidden 1):  50 neurons")
print("Layer 2 (Hidden 2):  50 neurons")
print("Layer 3 (Hidden 3):  50 neurons")
print("Layer 4 (Output):    50 neurons")

print("\nüìä a) INPUT MATRIX X:")
print("-" * 60)
print("Shape: (4, 10)")
print("Explanation:")
print("  ‚Ä¢ 4 rows = 4 instances (mini-batch size)")
print("  ‚Ä¢ 10 columns = 10 features (input neurons)")
print("  ‚Ä¢ Each row = one training example")

print("\nüìä b) HIDDEN LAYER WEIGHT MATRIX W_h:")
print("-" * 60)
print("For hidden layers, there are 3 weight matrices:")
print("  ‚Ä¢ W‚ÇÅ (Input ‚Üí Hidden1):   Shape (10, 50)")
print("  ‚Ä¢ W‚ÇÇ (Hidden1 ‚Üí Hidden2): Shape (50, 50)")
print("  ‚Ä¢ W‚ÇÉ (Hidden2 ‚Üí Hidden3): Shape (50, 50)")
print("\nGeneral rule: W_h has shape (n_prev, n_curr)")

print("\nüìä c) OUTPUT LAYER WEIGHT MATRIX W_o:")
print("-" * 60)
print("Shape: (50, 50)")
print("Explanation:")
print("  ‚Ä¢ 50 rows = neurons from Hidden Layer 3")
print("  ‚Ä¢ 50 columns = output neurons")

print("\nüìä d) OUTPUT MATRIX Y:")
print("-" * 60)
print("Shape: (4, 50)")
print("Explanation:")
print("  ‚Ä¢ 4 rows = 4 predictions (one per instance)")
print("  ‚Ä¢ 50 columns = 50 output values")

print("\nüìê e) EQUATION TO COMPUTE Y:")
print("-" * 60)
print("Forward pass equations:")
print()
print("H‚ÇÅ = œÉ(X @ W‚ÇÅ + b‚ÇÅ)        Shape: (4, 50)")
print("H‚ÇÇ = œÉ(H‚ÇÅ @ W‚ÇÇ + b‚ÇÇ)       Shape: (4, 50)")
print("H‚ÇÉ = œÉ(H‚ÇÇ @ W‚ÇÉ + b‚ÇÉ)       Shape: (4, 50)")
print("Y  = œÉ(H‚ÇÉ @ W_o + b_o)     Shape: (4, 50)")
print()
print("Where:")
print("  ‚Ä¢ @ = matrix multiplication")
print("  ‚Ä¢ œÉ = activation function (e.g., ReLU)")
print("  ‚Ä¢ b = bias vectors")

# Verify with actual computation
print("\nüî¨ VERIFICATION WITH NUMPY:")
print("-" * 60)

# Create random matrices
batch_size = 4
n_input = 10
n_hidden = 50
n_output = 50

X = np.random.randn(batch_size, n_input)
W1 = np.random.randn(n_input, n_hidden)
W2 = np.random.randn(n_hidden, n_hidden)
W3 = np.random.randn(n_hidden, n_hidden)
W_out = np.random.randn(n_hidden, n_output)

b1 = np.zeros(n_hidden)
b2 = np.zeros(n_hidden)
b3 = np.zeros(n_hidden)
b_out = np.zeros(n_output)

# Forward pass
H1 = np.maximum(0, X @ W1 + b1)  # ReLU
H2 = np.maximum(0, H1 @ W2 + b2)
H3 = np.maximum(0, H2 @ W3 + b3)
Y = np.maximum(0, H3 @ W_out + b_out)

print(f"X shape:    {X.shape}       ‚úì")
print(f"W1 shape:   {W1.shape}     ‚úì")
print(f"W2 shape:   {W2.shape}     ‚úì")
print(f"W3 shape:   {W3.shape}     ‚úì")
print(f"W_out shape: {W_out.shape}     ‚úì")
print(f"Y shape:    {Y.shape}      ‚úì")

print("\nüìà PARAMETER COUNT:")
print("-" * 60)
params_W1 = 10 * 50
params_W2 = 50 * 50
params_W3 = 50 * 50
params_Wout = 50 * 50
params_biases = 50 + 50 + 50 + 50

total_params = params_W1 + params_W2 + params_W3 + params_Wout + params_biases

print(f"W‚ÇÅ parameters:    {params_W1:,}")
print(f"W‚ÇÇ parameters:    {params_W2:,}")
print(f"W‚ÇÉ parameters:    {params_W3:,}")
print(f"W_out parameters: {params_Wout:,}")
print(f"Bias parameters:  {params_biases:,}")
print(f"-" * 60)
print(f"TOTAL PARAMETERS: {total_params:,}")

print("\n‚úÖ Exercise 6 completed!")
```

```python
# ============================================
# STEP 2: Build Deep MLP Architecture
# ============================================

print("\n" + "=" * 60)
print("STEP 2: BUILD DEEP MLP ARCHITECTURE")
print("=" * 60)

print("\nüèóÔ∏è Architecture Design:")
print("-" * 60)
print("Strategy: Deep network with dropout for regularization")

# Build model
model_mnist = keras.Sequential([
    # Input layer
    layers.Flatten(input_shape=[28, 28], name="input"),
    
    # Hidden layer 1
    layers.Dense(300, activation="relu", name="hidden1"),
    layers.Dropout(0.2, name="dropout1"),
    
    # Hidden layer 2
    layers.Dense(200, activation="relu", name="hidden2"),
    layers.Dropout(0.2, name="dropout2"),
    
    # Hidden layer 3
    layers.Dense(100, activation="relu", name="hidden3"),
    layers.Dropout(0.2, name="dropout3"),
    
    # Output layer
    layers.Dense(10, activation="softmax", name="output")
], name="deep_mnist_mlp")

print("\nüìä Model Architecture:")
model_mnist.summary()

print(f"\nüìà Total Parameters: {model_mnist.count_params():,}")
print("\n‚úÖ Deep MLP architecture built!")
```

```python
# ============================================
# STEP 3: Compile with Optimal Settings
# ============================================

print("\n" + "=" * 60)
print("STEP 3: COMPILE MODEL")
print("=" * 60)

# Compile with Adam optimizer (better than SGD for this task)
model_mnist.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

print("\n‚öôÔ∏è Compilation Settings:")
print("-" * 60)
print("  Optimizer: Adam")
print("  Learning Rate: 0.001")
print("  Loss: Sparse Categorical Cross-Entropy")
print("  Metrics: Accuracy")

print("\nüí° Why Adam?")
print("  ‚Ä¢ Adaptive learning rate per parameter")
print("  ‚Ä¢ Works well out-of-the-box")
print("  ‚Ä¢ Better than SGD for most tasks")

print("\n‚úÖ Model compiled!")
```

```python
# ============================================
# STEP 4: Setup Callbacks
# ============================================

print("\n" + "=" * 60)
print("STEP 4: SETUP CALLBACKS")
print("=" * 60)

import os
import time

# Create directories
os.makedirs("models", exist_ok=True)
log_dir = os.path.join("logs", "mnist", f"run_{int(time.time())}")
os.makedirs(log_dir, exist_ok=True)

# 1. ModelCheckpoint
checkpoint_cb = callbacks.ModelCheckpoint(
    filepath="models/best_mnist_model.keras",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max",
    verbose=1
)

# 2. EarlyStopping
early_stopping_cb = callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=10,
    mode="max",
    restore_best_weights=True,
    verbose=1
)

# 3. ReduceLROnPlateau
reduce_lr_cb = callbacks.ReduceLROnPlateau(
    monitor="val_accuracy",
    factor=0.5,
    patience=5,
    mode="max",
    min_lr=1e-7,
    verbose=1
)

# 4. TensorBoard
tensorboard_cb = callbacks.TensorBoard(
    log_dir=log_dir,
    histogram_freq=1,
    write_graph=True
)

print("\n‚úÖ Callbacks configured:")
print(f"  1. ModelCheckpoint ‚Üí models/best_mnist_model.keras")
print(f"  2. EarlyStopping (patience=10)")
print(f"  3. ReduceLROnPlateau (factor=0.5, patience=5)")
print(f"  4. TensorBoard ‚Üí {log_dir}")

print("\nüìä To view TensorBoard:")
print(f"  tensorboard --logdir={log_dir}")
```

```python
# ============================================
# STEP 5: Train Model
# ============================================

print("\n" + "=" * 60)
print("STEP 5: TRAIN MODEL")
print("=" * 60)

print("\nüöÄ Starting training...")
print("Target: > 98% test accuracy")
print("-" * 60)

# Train with all callbacks
history_mnist = model_mnist.fit(
    X_train, y_train,
    epochs=100,  # Will stop early if needed
    batch_size=128,
    validation_data=(X_valid, y_valid),
    callbacks=[
        checkpoint_cb,
        early_stopping_cb,
        reduce_lr_cb,
        tensorboard_cb
    ],
    verbose=1
)

print(f"\n‚úÖ Training completed!")
print(f"   Total epochs: {len(history_mnist.history['loss'])}")
```

```python
# ============================================
# STEP 6: Evaluate on Test Set
# ============================================

print("\n" + "=" * 60)
print("STEP 6: EVALUATE ON TEST SET")
print("=" * 60)

# Evaluate
test_loss, test_accuracy = model_mnist.evaluate(X_test, y_test, verbose=0)

print(f"\nüìà FINAL TEST RESULTS:")
print(f"  Test Loss: {test_loss:.4f}")
print(f"  Test Accuracy: {test_accuracy*100:.2f}%")

if test_accuracy >= 0.98:
    print(f"\nüéâ SUCCESS! Achieved > 98% accuracy!")
else:
    print(f"\n‚ö†Ô∏è Target not reached. Current: {test_accuracy*100:.2f}%")
    print(f"   Try: More epochs, different architecture, or hyperparameters")

# Visualize training history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

epochs_range = range(1, len(history_mnist.history['loss']) + 1)

# Plot 1: Loss
axes[0].plot(epochs_range, history_mnist.history['loss'], 'b-', linewidth=2, label='Training')
axes[0].plot(epochs_range, history_mnist.history['val_loss'], 'r-', linewidth=2, label='Validation')
axes[0].set_xlabel('Epoch', fontsize=12, weight='bold')
axes[0].set_ylabel('Loss', fontsize=12, weight='bold')
axes[0].set_title('Training & Validation Loss', fontsize=14, weight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: Accuracy
axes[1].plot(epochs_range, history_mnist.history['accuracy'], 'b-', linewidth=2, label='Training')
axes[1].plot(epochs_range, history_mnist.history['val_accuracy'], 'r-', linewidth=2, label='Validation')
axes[1].axhline(y=0.98, color='green', linestyle='--', linewidth=2, label='Target (98%)')
axes[1].set_xlabel('Epoch', fontsize=12, weight='bold')
axes[1].set_ylabel('Accuracy', fontsize=12, weight='bold')
axes[1].set_title('Training & Validation Accuracy', fontsize=14, weight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n‚úÖ Evaluation completed!")
```

```python
# ============================================
# STEP 7: Analyze Performance
# ============================================

print("\n" + "=" * 60)
print("STEP 7: PERFORMANCE ANALYSIS")
print("=" * 60)

# Make predictions
y_pred_proba = model_mnist.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)

# Confusion Matrix
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)

# Visualize confusion matrix
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax, cbar_kws={'label': 'Count'})
ax.set_xlabel('Predicted Label', fontsize=12, weight='bold')
ax.set_ylabel('True Label', fontsize=12, weight='bold')
ax.set_title(f'Confusion Matrix\nTest Accuracy: {test_accuracy*100:.2f}%', 
             fontsize=14, weight='bold')
plt.tight_layout()
plt.show()

# Classification report
print("\nüìä Classification Report:")
print("=" * 60)
print(classification_report(y_test, y_pred, 
                          target_names=[str(i) for i in range(10)]))

# Find misclassified examples
misclassified_idx = np.where(y_pred != y_test)[0]
print(f"\n‚ùå Misclassified: {len(misclassified_idx)} / {len(y_test)} images")
print(f"   Error rate: {len(misclassified_idx)/len(y_test)*100:.2f}%")

# Show some misclassified examples
if len(misclassified_idx) > 0:
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    for i, ax in enumerate(axes.flat):
        if i < len(misclassified_idx) and i < 10:
            idx = misclassified_idx[i]
            ax.imshow(X_test[idx], cmap='gray')
            ax.set_title(f'True: {y_test[idx]}, Pred: {y_pred[idx]}\n'
                        f'Conf: {y_pred_proba[idx][y_pred[idx]]*100:.1f}%',
                        fontsize=10, color='red')
            ax.axis('off')
    
    plt.suptitle('Misclassified Examples', fontsize=16, weight='bold', color='red')
    plt.tight_layout()
    plt.show()

print("\n‚úÖ Performance analysis completed!")
```
